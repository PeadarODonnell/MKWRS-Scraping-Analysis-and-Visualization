```python
import requests
from bs4 import BeautifulSoup
from numpy import asarray
from numpy import savetxt
import csv

URL = "https://mkwrs.com/mkwii/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")
```

```python
print(soup)
```

# get current and non-shortcut world records 
# (mario kart wii)

```python
results = soup.find_all("table", class_="wr")
current_world_records = results[0]
non_shortcut_world_records = results[1]
results = current_world_records
#print(results)


# split up current world records table into its rows.
# first row is the header, and the last row is the total time taken by all records.
#headings = results.find_all()
display_elements = results.find_all("tr")
#print(display_elements)
```

# get all headings for current and non shortcut world records

```python
# TESTING AREA
# HEADINGS

headings = display_elements[0].find_all("th")
#for point in data:
#    print(point.text.strip())
    
track = headings[0].text.strip()
time = headings[1].text.strip()
player = headings[2].text.strip()
mii_name = headings[3].text.strip()
nation = headings[4].text.strip()
date = headings[5].text.strip()
character = headings[6].text.strip()
vehicle = headings[7].text.strip()
controller = headings[8].text.strip()
duration = headings[9].text.strip()
ghost = headings[10].text.strip()

headings = [track,
            time,
            player,
            mii_name,
            nation,
            date,
            character,
            vehicle,
            controller,
            duration,
            ghost]
#print(headings)
```

# save all of the table data into arrays (EG track times, player names, vehicles, etc etc) and save to current_world_records_cv

```python
track_array = []
time_array = []
player_array = []
mii_name_array = []
nation_array = []
date_array = []
character_array = []
vehicle_array = []
controller_array = []
duration_array = []
ghost_array = []


for x in range(1, len(display_elements) - 1):
    data = display_elements[x].find_all("td")
    track_array.append(data[0].text.strip())
    time_array.append(data[1].text.strip())
    player_array.append(data[2].text.strip())
    mii_name_array.append((data[3].text.strip().encode("ascii", "ignore").decode("utf-8")))#.encode("utf-8"))
    nation_array.append(data[4].find('img').get('title'))
    date_array.append(data[5].text.strip())
    character_array.append(data[6].text.strip())
    vehicle_array.append(data[7].text.strip())
    controller_array.append(data[8].text.strip())
    duration_array.append(data[9].text.strip())
    ghost_array.append(data[10].find('a').get('href'))
    

all_current_world_records_table = [headings,
                                   track_array,
                                   time_array,
                                   player_array,
                                   mii_name_array,
                                   nation_array,
                                   date_array,
                                   character_array,
                                   vehicle_array,
                                   controller_array,
                                   duration_array,ghost_array]


all_current_world_record_contents = list(zip(*all_current_world_records_table[1:]))



f = open('Downloads/mario kart data/current_world_records_mario_kart_wii.csv', 'w', encoding="UTF8", newline="")
writer = csv.writer(f)
writer.writerow(headings)
for row in all_current_world_record_contents:
    #print(row)
    writer.writerow(row)
    
f.close()
```

# controller analysis

```python
gcn_controller_count = 0
wiimote_nunchuck_count = 0
wii_wheel_count = 0
classic_controller_count = 0

for i in controller_array:
    if(i == "GCN Controller"):
        gcn_controller_count += 1
    
    if(i == "Wiimote+Nunchuck"):
        wiimote_nunchuck_count += 1
    
    if(i == "Wii Wheel"):
        wii_wheel_count += 1
        
    if(i == "Classic Controller"):
        classic_controller_count += 1

print(gcn_controller_count +
     wiimote_nunchuck_count +
     wii_wheel_count +
     classic_controller_count)

controller_analysis = [gcn_controller_count,
                      wiimote_nunchuck_count,
                      wii_wheel_count,
                      classic_controller_count]
```

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [10, 6]
# Set up with a higher resolution screen (useful on Mac)
%config InlineBackend.figure_format = 'retina'
import pandas as pd
# Create a Pandas series from a list of values ("[]") and plot it:


plotdata = pd.DataFrame({"controller analysis": controller_analysis},
                       index=["GCN Controller",
                             "Wiimote+Nunchuck",
                             "Wii Wheel",
                             "Classic Controller"])

plotdata.plot(kind="barh", title="test")
#plt.xticks(rotation=30, horizontalalignment="center")
plt.title("controllers used for current mkwii world records")
plt.ylabel("controller")
plt.xlabel("world records")
plt.show()
```
